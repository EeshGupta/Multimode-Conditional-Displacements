# Q functions

```python
def plot_qfunc_2d(psi, xvec, yvec, label = ''):
    #fig, axes = plt.subplots(1, 2, subplot_kw={'projection': '3d'}, figsize=(12, 6))
    fig, ax = plt.subplots(1, 1, figsize=(6,5))
    Qf = qfunc(psi, xvec, xvec)
    pmesh = ax.pcolormesh(xvec, xvec, Qf, cmap='Blues')
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)
    plt.show()

    plt.show()
    
    
def plot_qfunc_2d_3d(psi, xvec, yvec, label = ''):
    X, Y = np.meshgrid(xvec, yvec)
    Z = qfunc(psi,xvec,yvec)
    
    fig = plt.figure(figsize = (12,5))
    ax = fig.add_subplot(122,projection = '3d')
    ax.contour3D(X, Y, Z, 50, cmap='Blues')
# ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
#                 cmap='Blues', edgecolor='none')
    ax.set_title(label);
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    ax.set_zlabel('Q');
    ax = fig.add_subplot(121)
    ax.set_title('Husimi Q function')
    Qf = qfunc(psi, xvec, xvec)
    pmesh = ax.pcolormesh(xvec, xvec, Qf, cmap='Blues')

    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)

    plt.show()
```

# Wigner functions

```python


def plot_wigner_2d(psi, xvec, yvec, label = ''):
    #fig, axes = plt.subplots(1, 2, subplot_kw={'projection': '3d'}, figsize=(12, 6))
    fig, ax = plt.subplots(1, 1, figsize=(6,5))
    Qf = wigner(psi, xvec, xvec)
    pmesh = ax.pcolormesh(xvec, xvec, Qf, cmap='Blues')
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)
    plt.show()

    plt.show()
    
    
def plot_wigner_2d_3d(psi, xvec, yvec, label = ''):
    X, Y = np.meshgrid(xvec, yvec)
    Z = wigner(psi,xvec,yvec)
    
    fig = plt.figure(figsize = (12,4.5))
    ax = fig.add_subplot(122,projection = '3d')
#     ax.contour3D(X, Y, Z, 50, cmap='Blues')
    ax.plot_surface(X, Y, Z, rstride=10, cstride=10,
                cmap='Blues', edgecolor='none')
    ax.set_title(label);
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    ax.set_zlabel('Q');
    ax = fig.add_subplot(121)
    ax.set_title('Wigner function')
    Qf = qfunc(psi, xvec, xvec)
    pmesh = ax.pcolormesh(xvec, xvec, Qf, cmap='Blues')

    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)

    plt.show()
```

```python
def plot_qfunc_wigner_fock(psi, xvec, yvec, label = ''):
    X, Y = np.meshgrid(xvec, yvec)
    Z = qfunc(psi,xvec,yvec)
    
    fig = plt.figure(figsize = (9,12))
    ax = fig.add_subplot(324,projection = '3d')
#     ax.contour3D(X, Y, Z, 50, cmap='Blues')
    ax.plot_surface(X, Y, Z, rstride=10, cstride=10,
                cmap='Blues', edgecolor='none')
    ax.set_title(label);
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    ax.set_zlabel('Q');
    ax = fig.add_subplot(323)
    ax.set_title('Husimi Q function')
    pmesh = ax.pcolormesh(xvec, xvec, Z, cmap='Blues')

    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)

    Z = wigner(psi,xvec,yvec)
    
    ax = fig.add_subplot(326,projection = '3d')
#     ax.contour3D(X, Y, Z, 50, cmap='RdBu')
    ax.plot_surface(X, Y, Z, rstride=10, cstride=10, vmin = -1/np.pi,vmax = 1/np.pi,
                cmap='RdBu', edgecolor='none')
    ax.set_title(label);
    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    ax.set_zlabel('$\\mathcal{W}$');
    ax = fig.add_subplot(325)
    ax.set_title('Wigner function')
    Qf = qfunc(psi, xvec, xvec)
    pmesh = ax.pcolormesh(xvec, xvec, Z,vmin = -1/np.pi,vmax = 1/np.pi,cmap='RdBu')

    ax.set_xlabel('Re[$\\alpha$]')
    ax.set_ylabel('Im[$\\alpha$]')
    fig.colorbar(pmesh, ax=ax)
    
    ax = fig.add_subplot(321)
    plot_fock_distribution(psi, fig=fig, ax=ax, title=label);
    
    plt.tight_layout()
    plt.show()
```

# Coherent states: $\left|\alpha\right>$

```python
psi = coherent(N, 2)
xvec = np.linspace(-8, 8, 500)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = '$\\alpha = 2$')
```

## Superposition of coherent states

```python
psi = (coherent(N, -2.0) + coherent(N, 2.0)) / np.sqrt(2)
xvec = np.linspace(-5, 5, 500)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of coherent states')
```

```python
psi = (coherent(N, -2.0*(1+1j)/np.sqrt(2)) + coherent(N, 2.0*(1+1j)/np.sqrt(2))) / np.sqrt(2)
xvec = np.linspace(-5, 5, 500)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of coherent states')
```

```python
psi = (coherent(N, -2.0) + coherent(N, -2j) + coherent(N, 2j) + coherent(N, 2.0)).unit()
xvec = np.linspace(-5, 5, 500)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of 4 coherent states')
```

```python
psi = (coherent(N, -2.0) + coherent(N, -1j) + coherent(N, 1j) + coherent(N, 2.0)).unit()
xvec = np.linspace(-5, 5, 500)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of 4 coherent states')
```

```python
NN = 3
xvec = np.linspace(-5, 5, 500)
for n in range(NN):
    psi = sum([coherent(N, 2*np.exp(2j * np.pi * m / (n + 2))) for m in range(n + 2)]).unit()

    plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Coherent state superposition with n = '+str(n + 2))

    #if n < NN - 1:
    #    axes[n].set_ylabel("")
```

### Mixture of coherent states

# Squeezed states 

## Squeezed vacuum states

```python
psi = squeeze(N, 0.75) * basis(N, 0)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Squeezed Vacuum')

psi = squeeze(N, 0.75j) * basis(N, 0)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Squeezed Vacuum')
```

## Superposition of squeezed vacuum

```python
psi = (squeeze(N, 0.75j) * basis(N, 0) - squeeze(N, -0.75j) * basis(N, 0)).unit()
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of squeezed vacuum')
```

```python
psi = squeeze(N, 0.75j) * basis(N, 1)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Squeezed Fock 1')
```

## Displaced squeezed vacuum

```python
psi = displace(N, 2) * squeeze(N, 0.75*(1+1j)/np.sqrt(2)) * basis(N, 0)
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Displaced squeezed vacuum')
```

## Superposition of two displaced squeezed states

```python
psi = (displace(N, -1) * squeeze(N, 0.75) * basis(N, 0) - displace(N, 1) * squeeze(N, -0.75) * basis(N, 0)).unit()
plot_qfunc_wigner_fock(psi,xvec,xvec, label = 'Superposition of two displaced squeezed states')
```
