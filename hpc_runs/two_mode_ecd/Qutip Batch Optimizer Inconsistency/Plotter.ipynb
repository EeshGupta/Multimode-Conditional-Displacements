```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
filename = "Qutip_Inconsistency.csv"
df1 = pd.read_csv(filename)
df1 = df1.drop(0)
df1
```

```python
filename = "Qutip_Inconsistency2.csv"
df2 = pd.read_csv(filename)
df2 = df2.drop(0)
df2
```

```python
filename = "Qutip_Inconsistency3.csv"
df3 = pd.read_csv(filename)
df3 = df3.drop(0)
df3
```

```python
frames = [df1, df2, df3]
df = pd.concat(frames)
df = df.reset_index()
df
```

```python
df['diffs'] =  df['exact fidelity'] - df['qutip fidelity']
df
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8, 4))
x = df['truncation']
y = df['levels']
c = df['diffs']
plt.scatter(x,y,c = c, cmap= "viridis")
cbar = plt.colorbar()
cbar.set_label('True- Qutip Fidelity')
plt.xlabel('Truncation')
plt.ylabel('Layers')
plt.title(r'$g01 \rightarrow g10$ Qutip-Exact fidelity mismatch', fontsize = 15)
#fig.savefig('Layer vs Fidelity for g01g02_to_g10_g20(no outlier)')
plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8, 4))
x = df['truncation']
y = df['levels']
c = df['exact fidelity']
plt.scatter(x,y,c = c, cmap= "plasma")
cbar = plt.colorbar()
cbar.set_label('Exact fidelity')
plt.xlabel('Truncation')
plt.ylabel('Layers')
plt.title(r'$g01 \rightarrow g10$ Exact fidelity', fontsize = 15)
#fig.savefig('Layer vs Fidelity for g01g02_to_g10_g20(no outlier)')
plt.tight_layout()
```

```python
df['diffs2'] =  df['bo_fidelity'] - df['exact fidelity']

fig = plt.figure(figsize = (8, 4))
x = df['truncation']
y = df['levels']
c = df['diffs2']
plt.scatter(x,y,c = c, cmap= "inferno")
cbar = plt.colorbar()
cbar.set_label('(batch opt - exact fidelity) BCH error')
plt.xlabel('Truncation')
plt.ylabel('Layers')
plt.title(r'$g01 \rightarrow g10$ Baker Campbell Haussdoff error', fontsize = 15)
#fig.savefig('Layer vs Fidelity for g01g02_to_g10_g20(no outlier)')
plt.tight_layout()
```

```python
conditions = ['bo_fidelity', 'only ecd', 'bare_qubit_mode_coupling','Stark Shift', 'Mode Coupling', 'transmon_relax','transmon_dephasing']
x_pos = np.arange(len(conditions))
bars = []
bars_err = []
for i in range(len(conditions)):
    listy = list(df[conditions[i]]) 
    bars.append(1 - np.mean(listy))
    bars_err.append(np.std(listy))

fig, ax = plt.subplots(figsize = (8,6))
ax.bar(conditions, bars, yerr=bars_err, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('$1-\mathcal{F}$')
#ax.set_xticks(x_pos)
ax.set_xticklabels(conditions, fontsize = 15)
ax.set_title('Qutip Noise Simulation of g01 -> g10')
ax.yaxis.grid(True)
plt.xticks(range(len(conditions)), conditions, rotation = 90)
```
